```python
## 조회수 전처리 함수
def preprocess_views(data, columns):
  for i in range(len(data)):
    for c in columns:
      data.iloc[i, c] = data.iloc[i, c].replace(',', '')  # , 제거
      if '만' in data.iloc[i, c]:
        data.iloc[i, c] = int(data.iloc[i, c][:-1]) * 10000  # 만 제거 및 int형 변환
  data = data.astype({'1화 조회수' : 'int64', '2화 조회수' : 'int64', '3화 조회수' : 'int64', '4화 조회수' : 'int64', '5화 조회수' : 'int64'})
  return data
```

### 본문 글자수 정보

#### best novel

```python
sample = b_data.iloc[0, 4]
sample = sample.replace('_x000D_', '')
print(sample)
```

```python
# 글자수 계산 함수
def count_blank(data):
  text = data
  split_data = text.split()
  word = len(split_data)  # word = 단어 수, word - 1 = 공백 수

  count = len(text)  # count = 공백 포함 글자수, count - word + 1 = 공백 제거 글자수

  return(word, word - 1, count, count - word + 1)

count_blank(sample)
```

```python
# 본문 데이터 불러오기
content = b_data.iloc[:, [4, 6, 8, 10, 12]]

for i in range(5):
  for j in range(140):
    content.iloc[j, i] = content.iloc[j, i].replace('_x000D_', '')  # 불필요한 토큰 삭제
content
```

```python
for i in range(1, 6):
  for c in ['단어 수', '공백 수', '공백 포함 글자수', '공백 제거 글자수']:
    col = '%d화 %s' %(i, c)
    content[col] = 0

content
```

```python
content.info()
```

```python
for i in range(5):
  for j in range(140): 
    word, blank, char_blank, char_no_blank = count_blank(content.iloc[j, i])

    start_num = 5 + i * 4

    content.iloc[j, start_num] = int(word)
    content.iloc[j, start_num+1] = int(blank)
    content.iloc[j, start_num+2] = int(char_blank)
    content.iloc[j, start_num+3] = int(char_no_blank)

content.head(5)
```
